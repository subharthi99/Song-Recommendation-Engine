Importing Modules

```python
import pandas as pd
import numpy as np
#custom Python class
import Recommenders as Recommenders
```

Dataset Loading

```python
song_data_1 = pd.read_csv('triplets_file.csv')
song_data_1.head()
```

```python
song_data_2 = pd.read_csv('song_data.csv')
song_data_2.head()
```

Combining both the dataframes

```python
song_data = pd.merge(song_data_1, song_data_2.drop_duplicates(['song_id']), on='song_id', how='left')
song_data.head()
```

Data Preprocessing

```python
# Combining title and artist name
song_data['song'] = song_data['title']+' - '+song_data['artist_name']
song_data.head()
```

```python
song_data = song_data.head(10000)
# cummulative sum of listen count of the songs
song_grouped = song_data.groupby(['song']).agg({'listen_count':'count'}).reset_index()
song_grouped.head()
```

```python
grouped_sum = song_grouped['listen_count'].sum()
song_grouped['percentage'] = (song_grouped['listen_count'] / grouped_sum ) * 100
song_grouped.sort_values(['listen_count', 'song'], ascending=[0,1])
```

Popularity Based Recommendation Engine

```python
ppl = Recommenders.popularity_recommender_py()
```

```python
ppl.create(song_data, 'user_id', 'song')
```

```python
# display the top 10 popular songs
ppl.recommend(song_data['user_id'][5])
```

```python
ppl.recommend(song_data['user_id'][100])
```

Item Similarity Based Recommendation Engine

```python
isim = Recommenders.item_similarity_recommender_py()
isim.create(song_data, 'user_id', 'song')
```

```python
user_items = isim.get_user_items(song_data['user_id'][5])
```

```python
# display user songs history
for user_item in user_items:
    print(user_item)
```

```python
# song recommendation for that user
isim.recommend(song_data['user_id'][5])
```
